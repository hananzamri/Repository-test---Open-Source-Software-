```python
print("="*40)
project = ("NexusOS", "1.0", 2026,"Python", "Hanan" )
print(project[0], " - Open Source Project")
print("Project Lead : ", project[4])
print("="*40)

print("Name :", project[0], "| Version :", project[1], "| Started :", project[2])

print("First 3 fields :", project[0:3])

print("Language count :", project.count("Python"))
print("Language index :", project.index("Python"))

# project[0] = "test"
# TypeError: tuple does not support item assignment
# Tuples are immutable, making them suitable for fixed, official project data.

print("-"*40)

contributors = []

for i in range(4):
    print("Enter contributor", i+1)
    contributor= {
        "name" : input("Name      :"),
        "role" : input("Role      :"),
        "language" : input("Language  :"),
        "commits" : input("Commits   :"),
        "country" : input("Country   :"),
    }
    contributors.append(contributor)
    print("-"*40)

names = [c["name"] for c in contributors]
names.sort()

print("Sorted names :", names)
print("Last name  : ", names[-1], "  First two : ", names[:2])

for c in contributors:
    c.update({"status":"Active"})

print("Contributor 1 status :", contributors[0].get("status"))

backup = contributors[0].copy()
print("Backup :", backup)



print("-"*40)

issues =[]


for i in range (5):
  print("ISSUE", i+1)
  issue = {
      "id" : input("ID                                    : "),
      "title": input("Title                                 : "),
      "type": input("Type (Bug/Feature)                    : "),
      "priority": input("Priority (Critical/High/Medium/Low)   : "),
      "reporter": input("Reporter                              : "),
      "status": input("Status (Open/In Progress/Resolved)    : ")
  }
  issues.append(issue)
  print("-"*40)


print("-"*40)
open_count = 0
for issue in issues:
    if issue["status"] == "Open":
        open_count += 1

print("Open issues :", open_count)
issues[0]["priority"] = "Critical"
print("First issue → priority updated to Critical.")
print("Last two issues :", issues[-2:])

print("-"*40)

reporters = set()
for issue in issues:
    reporters.add(issue["reporter"])

tech_stack = set()
for c in contributors:
    tech_stack.add(c["language"])

tech_stack.add("Go")
tech_stack.discard("Rust")

#union_set = reporters.union(tech_stack)
#intersection_set = reporters.intersection(tech_stack)
#difference_set = reporters.difference(tech_stack)

print("-"*40)
print("reporters :", reporters)
print("tech_stack :", tech_stack)
print("union :", reporters.union(tech_stack))
print("intersection :", reporters.intersection(tech_stack), "difference :", reporters.difference(tech_stack)
)

priority_set = set()
for issue in issues:
    priority_set.add(issue["priority"])

if "Critical" in priority_set:
    print("Critical present : YES — flag for immediate review.")
else:
    print("Critical present : NO")

print("-"*40)

priority_count = {}
for issue in issues:
    p = issue["priority"]
    if p in priority_count:
        priority_count[p] += 1
    else:
        priority_count[p] = 1

status_groups = {}
for issue in issues:
    s = issue["status"]
    if s not in status_groups:
        status_groups[s] = []
    status_groups[s].append(issue["title"])

print("-"*40)
print("priority keys :", priority_count.keys())
print("priority values :", priority_count.values())


for key, value in status_groups.items():
    print("Status Groups →", key, ":", value)

report_count = {}
for issue in issues:
    r = issue["reporter"]
    if r in report_count:
        report_count[r] += 1
    else:
        report_count[r] = 1

top_reporter = ""
max_count = 0

for r in report_count:
    if report_count[r] > max_count:
        max_count = report_count[r]
        top_reporter = r

print("Top reporter :", top_reporter, "(", max_count, "issues )")

# pop() remove 'type'
removed = issues[0].pop("type")
print("After pop('type') :", issues[0])

import os

print("-"*40)

# Create folder name from project tuple
folder_name = project[0].lower().replace(" ", "_")

if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print("Folder created :", folder_name)
else:
    print("Folder already exists :", folder_name)

print("Folder contents :", os.listdir(folder_name))

# File paths
report_path = os.path.join(folder_name, "project_report.txt")
csv_path = os.path.join(folder_name, "issues.csv")


try:
    # WRITE REPORT
    with open(report_path, "w") as f:
        f.write("="*40 + "\n")
        f.write(project[0] + " — PROJECT REPORT\n")
        f.write("="*40 + "\n")

        f.write("Project Info:\n")
        f.write(str(project) + "\n\n")

        f.write("Contributors:\n")
        for c in contributors:
            f.write(str(c) + "\n")

        f.write("\nIssues:\n")
        for i in issues:
            f.write(str(i) + "\n")

        f.write("\nAnalysis:\n")
        f.write("Open issues : " + str(open_count) + "\n")
        f.write("Top reporter : " + top_reporter + "\n")
        f.write("Priority count : " + str(priority_count) + "\n")

    # WRITE CSV
    with open(csv_path, "w") as f:
        f.write("id,title,priority,reporter,status\n")
        for i in issues:
            f.write(f"{i['id']},{i['title']},{i['priority']},{i['reporter']},{i['status']}\n")

    print("Files saved :", os.listdir(folder_name))

except IOError:
    print("Error writing files")

try:
    print("--- read() ---")
    with open(report_path, "r") as f:
        print(f.read())

    print("--- readline() ---")
    with open(report_path, "r") as f:
        print("Line 1 :", f.readline().strip())
        print("Line 2 :", f.readline().strip())

    print("--- readlines() ---")
    with open(report_path, "r") as f:
        lines = f.readlines()
        print("Total lines :", len(lines))

        count = 0
        for line in lines:
            if "Critical" in line or "High" in line:
                count += 1
                print(line.strip())

        print("Critical/High lines :", count)

except FileNotFoundError:
    print("File not found")

print("="*40)
print(f"{project[0]} — FINAL SUMMARY")
print("="*40)

print(f"Project : {project[0]} Version : {project[1]} Lead : {project[4]}")
print(f"Contributors : {len(contributors)} Names : {names}")
print(f"Tech Stack : {tech_stack}")
print(f"Issues : {len(issues)} Open : {open_count} Reporters : {len(reporters)}")
print(f"Top Reporter : {top_reporter} ({max_count} issues)")

# Priority breakdown
priority_text = ""
for k, v in priority_count.items():
    priority_text += f"{k}:{v} "

print(priority_text.strip())

print(f"Report : {report_path}")
print(f"CSV : {csv_path}")

print("="*40)
print(f"{project[0]} complete. Thank you for contributing to open source!")
print("="*40)
```
